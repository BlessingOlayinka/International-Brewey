# 1. Distribution by Race/Ethnicity:


###### - What is the racial/ethnic composition of the workforce?

```python
# to get the Racial composition we are going to do value counts
# Count the number of employees for each racial/ethnic group
# This code will calculate the percentage of employees in each racial/ethnic group and display the result.

race_composition = round(df1['RACE'].value_counts(normalize = True) * 100, 2)
print(race_composition)
```

###### - Are there any significant disparities in the representation of different racial/ethnic groups across departments or positions?

```python
# Calculate the percentage of each racial/ethnic group in each department
racial_composition_department = df1.groupby(['DEPARTMENT', 'RACE']).size().unstack(fill_value=0)
racial_composition_department = racial_composition_department.div(racial_composition_department.sum(axis = 1), axis = 0) * 100

# Calculate the percentage of each racial/ethnic group in each position
racial_composition_position = df1.groupby(['POSITION_TITLE', 'RACE']).size().unstack(fill_value=0)
racial_composition_position = racial_composition_position.div(racial_composition_position.sum(axis=1), axis=0) * 100

# Displaying the result
print("Racial/ethnic composition by department:")
print(racial_composition_department)
print("\nRacial/ethnic composition by position:")
print(racial_composition_position)
```

# 2. Gender Representation:

###### - What is the gender distribution within the organization?

```python
# Count the number of employees for each gender
gender_distribution = df1['GENDER'].value_counts(normalize = True) * 100
print(gender_distribution)
```

###### - Are there any gender imbalances in certain departments or roles?


```python
# Calculate the percentage of each gender in each department
gender_distribution_department = df1.groupby(['DEPARTMENT', 'GENDER']).size().unstack(fill_value=0)
gender_distribution_department = round(gender_distribution_department.div(gender_distribution_department.sum(axis=1), axis=0) * 100, 2)

# Calculate the percentage of each gender in each position
gender_distribution_position = df1.groupby(['POSITION_TITLE', 'GENDER']).size().unstack(fill_value=0)
gender_distribution_position = gender_distribution_position.div(gender_distribution_position.sum(axis=1), axis=0) * 100

# Display the results
print("Gender distribution by department:")
print(gender_distribution_department)

print("\nGender distribution by position:")
print(gender_distribution_position)
```

# 3. Employment Type Distribution:

###### - What proportion of employees are full-time, part-time, or contract workers?


```python
# Calculate the proportion of each employment type
Emplo_type_distribution = df1['EMPLOYMENT_TYPE'].value_counts(normalize = True) * 100
print("The Proportion of each employment type is down below:\n")
print(Emplo_type_distribution)
```

###### - Are there differences in employment types across departments or positions?

```python
# Calculate the proportion of each employment type in each department
employment_type_depart = df1.groupby(['DEPARTMENT', 'EMPLOYMENT_TYPE']).size().unstack(fill_value=0)
employment_type_depart = round(employment_type_depart.div(employment_type_depart.sum(axis=1), axis=0) * 100, 2)

# Calculate the proportion of each employment type in each position
employment_type_positions = df1.groupby(['POSITION_TITLE', 'EMPLOYMENT_TYPE']).size().unstack(fill_value=0)
employment_type_positions = employment_type_positions.div(employment_type_positions.sum(axis=1), axis=0) * 100

print("Employment type distribution by department:")
print(employment_type_depart)

print("\nEmployment type distribution by position:")
print(employment_type_positions)
```

# 6. Salary Distribution:


###### - What is the distribution of base salaries across the organization? (mean, mode, median)

```python

# Calculate the mean, mode, and median base salaries
mean_salary = df1['BASE_SALARY'].mean()
mode_salary = df1['BASE_SALARY'].mode().iloc[0]  # Mode can have multiple values, so we take the first one
median_salary = df1['BASE_SALARY'].median()

# Display the results
print("Mean base salary:", mean_salary)
print("Mode base salary:", mode_salary)
print("Median base salary:", median_salary)
```

# 7. Salary Disparities:

###### - Are there significant differences in salaries between different demographic groups (e.g., race, gender)?

```python
from scipy.stats import ttest_ind

# Split the dataset by gender
male_salaries = df1[df1['GENDER'] == 'Male']['BASE_SALARY']
female_salaries = df1[df1['GENDER'] == 'Female']['BASE_SALARY']

# Perform a t-test
t_stat, p_value = ttest_ind(male_salaries, female_salaries, equal_var=False)

# Display the result
if p_value < 0.05:
    print("There is a significant difference in salaries between male and female employees.")
else:
    print("There is no significant difference in salaries between male and female employees.")
```

```python
from scipy.stats import f_oneway

# Create a list of salaries for each racial/ethnic group
racial_groups = df1['RACE'].unique()
group_salaries = [df1[df1['RACE'] == race]['BASE_SALARY'] for race in racial_groups]

# Perform ANOVA test
f_stat, p_value = f_oneway(*group_salaries)

# Display the result
if p_value < 0.05:
    print("There is a significant difference in salaries between different racial/ethnic groups.")
else:
    print("There is no significant difference in salaries between different racial/ethnic groups.")
```

###### - Is there evidence of gender or racial pay gaps within the organization?

```python

# Split the dataset by gender
male_salaries = df1[df1['GENDER'] == 'Male']['BASE_SALARY']
female_salaries = df1[df1['GENDER'] == 'Female']['BASE_SALARY']

# Perform t-test
t_stat, p_value = ttest_ind(male_salaries, female_salaries, equal_var=False)

# Display the result
if p_value < 0.05:
    print("There is evidence of a gender pay gap within the organization.")
else:
    print("There is no evidence of a gender pay gap within the organization.")
```

# 8. Salary by Department/Position:

###### - What is the average salary within each department or position title?

```python
# Calculate the average salary within each department
average_salary_by_department = round(df1.groupby('DEPARTMENT')['BASE_SALARY'].mean(), 2)

# Calculate the average salary within each position title
average_salary_by_position = round(df1.groupby('POSITION_TITLE')['BASE_SALARY'].mean(), 2)

print('\n Average salary woithin each department: \n')
print(average_salary_by_department)

print('\n Average salary woithin each positions: \n')
print(average_salary_by_position)
```

###### - Are there departments or roles that offer higher salaries compared to others?

```python
# Calculate the average salary within each department
average_salary_department = df1.groupby('DEPARTMENT')['BASE_SALARY'].mean()

# Find the department with the highest average salary
highest_salary_department = average_salary_department.idxmax()
highest_average_salary_department = round(average_salary_department.max(),2)

# Display the result
print("Department with the highest average salary:", highest_salary_department)
print("Highest average salary within a department:", highest_average_salary_department)
```

```python
# Calculate the average salary within each position title
average_salary_position = df1.groupby('POSITION_TITLE')['BASE_SALARY'].mean()

# Find the position title with the highest average salary
highest_salary_position = average_salary_position.idxmax()
highest_average_salary_position = average_salary_position.max()

# Display the result
print("Position title with the highest average salary:", highest_salary_position)
print("Highest average salary within a position title:", highest_average_salary_position)
```
